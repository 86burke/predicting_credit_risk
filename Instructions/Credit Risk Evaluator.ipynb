```python
import numpy as np
import pandas as pd
from pathlib import Path
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler
```

```python
train_df = pd.read_csv(Path('Resources/2019loans.csv'))
test_df = pd.read_csv(Path('Resources/2020Q1loans.csv'))
```

```python
# combining dataframes
combined=pd.concat([train_df,test_df])
combined
```

```python
# Convert categorical data to numeric and separate target feature for training data
cnvt_num=pd.get_dummies(combined)
cnvt_num.columns
```

```python
# Convert categorical data to numeric and separate target feature for testing data
X=cnvt_num.drop(columns='loan_status_low_risk')
y=cnvt_num['loan_status_low_risk']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=38)
```

```python
# Train the Logistic Regression model on the unscaled data and print the model score
clf = LogisticRegression(random_state=38,max_iter=1200000).fit(X_train, y_train)
clf_acc=clf.score(X_test,y_test)
print(f"Trained Data Score: {clf_acc}")
```

```python
# Train a Random Forest Classifier model and print the model score
rfclf = RandomForestClassifier(max_depth=2, random_state=0).fit(X_train, y_train)
rfclf_acc=rfclf.score(X_test,y_test)
print(f"Trained Data Score: {rfclf_acc}")
```

```python
# Scale the data
X_train_scaled=StandardScaler().fit_transform(X_train)
X_test_scaled=StandardScaler().fit_transform(X_test)
```

```python
# Train the Logistic Regression model on the scaled data and print the model score
clf_scaled = LogisticRegression(random_state=38,max_iter=1200000).fit(X_train_scaled, y_train)
clf_acc_scaled=clf.score(X_test_scaled,y_test)
print(f"Trained Data Score: {clf_acc_scaled}")
```

```python
# Train a Random Forest Classifier model on the scaled data and print the model score
rfclf = RandomForestClassifier(max_depth=2, random_state=0).fit(X_train_scaled, y_train)
rfclf_acc=rfclf.score(X_test_scaled,y_test)
print(f"Trained Data Score: {rfclf_acc}")
```
